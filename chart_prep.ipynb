```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Dataset was created in Excel by taking the modulus of the row number (=MOD(ROW(A1),4)=1)
iris = pd.read_csv("iris_subset.csv")
```

```python
iris
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
plot = sns.scatterplot(x=iris['Petal_Length'], y=iris['Petal_Width'], hue=iris['Class'], s=100)
plot.get_figure().savefig("petal_length_versus_width.png")
```
